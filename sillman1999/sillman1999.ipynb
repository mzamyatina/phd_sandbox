```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Files are created with Engauge-digitizer
x1 = np.genfromtxt('O3plot_right_axis_c1.csv', delimiter=',', skip_header=1, usecols=(0))
y1 = np.genfromtxt('O3plot_right_axis_c1.csv', delimiter=',', skip_header=1, usecols=(1))
x2 = np.genfromtxt('O3plot_right_axis_c2.csv', delimiter=',', skip_header=1, usecols=(0))
y2 = np.genfromtxt('O3plot_right_axis_c2.csv', delimiter=',', skip_header=1, usecols=(1))
x3 = np.genfromtxt('O3plot_right_axis_c3.csv', delimiter=',', skip_header=1, usecols=(0))
y3 = np.genfromtxt('O3plot_right_axis_c3.csv', delimiter=',', skip_header=1, usecols=(1))
x4 = np.genfromtxt('O3plot_right_axis_c4.csv', delimiter=',', skip_header=1, usecols=(0))
y4 = np.genfromtxt('O3plot_right_axis_c4.csv', delimiter=',', skip_header=1, usecols=(1))
x5 = np.genfromtxt('O3plot_right_axis_c5.csv', delimiter=',', skip_header=1, usecols=(0))
y5 = np.genfromtxt('O3plot_right_axis_c5.csv', delimiter=',', skip_header=1, usecols=(1))
x6 = np.genfromtxt('O3plot_right_axis_c6.csv', delimiter=',', skip_header=1, usecols=(0))
y6 = np.genfromtxt('O3plot_right_axis_c6.csv', delimiter=',', skip_header=1, usecols=(1))
x7 = np.genfromtxt('O3plot_right_axis_c7.csv', delimiter=',', skip_header=1, usecols=(0))
y7 = np.genfromtxt('O3plot_right_axis_c7.csv', delimiter=',', skip_header=1, usecols=(1))
```

```python
plt.rcParams['mathtext.default'] = 'regular'
line_kwargs = dict(color='k', linestyle='solid')
fontsize = 18

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x1, y1, **line_kwargs)
ax.plot(x2, y2, **line_kwargs)
ax.plot(x3, y3, **line_kwargs)
ax.plot(x4, y4, **line_kwargs)
ax.plot(x5, y5, **line_kwargs)
ax.plot(x6, y6, **line_kwargs)
ax.plot(x7, y7, **line_kwargs)
ax.set_xlim(1, 100)
ax.set_ylim(10,1000)
ax.set_xlabel('$NO_x$, ppb', fontsize=fontsize)
ax.set_ylabel('VOC, ppbC', fontsize=fontsize);

xticklabels = [1, 2, 5, 10, 20, 50, 100]
xticks = [1, 15, 35, 50, 65, 85, 100]
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, fontsize=fontsize)

yticklabels = [10, 100, 1000]
yticks = [10, 500, 1000]
ax.set_yticks(yticks)
ax.set_yticklabels(yticklabels, fontsize=fontsize);

# 1, 2.5, 5, 10, 15, 20 and 30 ppb/h
fig.text(0.54, 0.37, '1', backgroundcolor='w', rotation=25, fontsize=16)
fig.text(0.52, 0.44, '2.5', backgroundcolor='w', rotation=25, fontsize=16)
fig.text(0.54, 0.52, '5', backgroundcolor='w', rotation=15, fontsize=16)
fig.text(0.53, 0.60, '10', backgroundcolor='w', rotation=10, fontsize=16)
fig.text(0.53, 0.66, '15', backgroundcolor='w', rotation=6, fontsize=16)
fig.text(0.53, 0.705, '20', backgroundcolor='w', rotation=4, fontsize=16)
fig.text(0.53, 0.78, '30', backgroundcolor='w', rotation=-2, fontsize=16);

fig.savefig('sillman1999_reconstructed_flipped.png', bbox_inches='tight', dpi=300)
```
