```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
plt.rcParams['mathtext.default'] = 'regular'
```

```python
savefig = True
path_to_figs = Path.home() / 'UEA' / 'PhD' / 'results' / 'misc'
```

```python
fs = 16
props_vars = dict(boxstyle='round', facecolor='w', alpha=1)
props_chem = dict(boxstyle='circle',  facecolor='w', linestyle='dashed', linewidth=5, alpha=1)
fig, ax = plt.subplots(figsize=(7.1,5), facecolor='w')
ax.plot()
ax.set_xlim(0,10)
ax.set_ylim(0,7)

ax.text(3,3.55,'Chemical reactions', fontsize=fs, bbox=props_chem)

ax.text(0.8,5,'Initialised\nvariable species\n$O_3$, NO', fontsize=fs, bbox=props_vars)
ax.text(6,5,'Fixed species\nCO, $C_1$-$C_3$ alkanes', fontsize=fs, bbox=props_vars)
ax.text(1,2.2,'Emission\nNO, $NO_2$', fontsize=fs)
ax.text(6.9,2.2,'Deposition\n$O_3$, $H_2O_2$, $HNO_3$', fontsize=15)
ax.arrow(3,0,0,2, width=0.1, length_includes_head=True, fc='k', ec='k')
ax.arrow(7,2,0,-2, width=0.1, length_includes_head=True, fc='k', ec='k')
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
if savefig: fig.savefig(path_to_figs / 'scheme_box_model_C1C3.png', bbox_inches='tight', dpi=200)
```
