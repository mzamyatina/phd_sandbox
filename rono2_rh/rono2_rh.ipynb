```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['mathtext.default'] = 'regular'
```

```python
# Read data
file_name = 'ENV-MA80_AlkylNitratesExercise_solution.xlsx'
dfs = pd.read_excel(file_name, sheet_name=None)
```

```python
# Rename columns
dfs['NAMBLEXobs'].rename(columns={'Date_Julian_Day':'JulianDay',
                                  'EtNO3_ppt':'EtONO2', '1PrNO3_ppt':'nPrONO2', '2PrNO3_ppt':'iPrONO2', '2BuNO3_ppt':'sBuONO2',
                                  'ethane_ppt':'C2H6', 'propane_ppt':'C3H8', 'n-butane_ppt':'nC4H10'}, inplace=True)
dfs['ITOPobs'].rename(columns={'MeNO3_ppt':'MeONO2', 'EtNO3_ppt':'EtONO2',
                               '1PrNO3_ppt':'nPrONO2', '2PrNO3_ppt':'iPrONO2', '2BuNO3_ppt':'sBuONO2',
                               '2+3PeNO3_ppt':'nPeBCONO2', '3Me2BuNO3_ppt':'iPeBONO2',
                               'ethane_ppt':'C2H6', 'propane_ppt':'C3H8', 'n-butane_ppt':'nC4H10',
                               'n-pentane_ppt':'nC5H12', 'i-pentane_ppt':'iC5H12'}, inplace=True)
dfs['TORCH1obs'].rename(columns={'Time_Julian_Day':'JulianDay', 'EtNO3_ppt':'EtONO2', '1PrNO3_ppt':'nPrONO2', '2PrNO3_ppt':'iPrONO2', 
                                 '1BuNO3_ppt':'nBuONO2', '2BuNO3_ppt':'sBuONO2', '2+3PeNO3_ppt':'nPeBCONO2', 
                                 '3Me2BuNO3_ppt':'iPeBONO2', 
                                 'ethane_ppt':'C2H6', 'propane_ppt':'C3H8', 'nbutane_ppt':'nC4H10',
                                 'n-pentane_ppt':'nC5H12', 'i-pentane_ppt':'iC5H12'}, inplace=True)
dfs['TORCH2obs'].rename(columns={'Time_Julian_Day':'JulianDay', 'EtNO3_ppt':'EtONO2', '1PrNO3_ppt':'nPrONO2', '2PrNO3_ppt':'iPrONO2', 
                                 '1BuNO3_ppt':'nBuONO2', '2BuNO3_ppt':'sBuONO2', '2+3PeNO3_ppt':'nPeBCONO2', 
                                 '3Me2BuNO3_ppt':'iPeBONO2', 
                                 'ethane_ppt':'C2H6', 'propane_ppt':'C3H8', 'n-butane_ppt':'nC4H10',
                                 'n-pentane_ppt':'nC5H12', 'i-pentane_ppt':'iC5H12'}, inplace=True)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(dfs['NAMBLEXobs'].C2H6, color='b')
ax2 = ax1.twinx()
plt.plot(dfs['NAMBLEXobs'].EtONO2, color='r')
```

```python
NAMBLEX_kwargs = dict(label='NAMBLEX', color='C0')
ITOP_kwargs = dict(label='ITOP', color='C1')
TORCH1_kwargs = dict(label='TORCH1', color='C2')
TORCH2_kwargs = dict(label='TORCH2', color='C3')

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 7))#, sharex=True, sharey=True)
ax[0,0].scatter(dfs['NAMBLEXobs'].EtONO2/dfs['NAMBLEXobs'].C2H6, dfs['NAMBLEXobs'].sBuONO2/dfs['NAMBLEXobs'].nC4H10, **NAMBLEX_kwargs)
ax[0,0].scatter(dfs['ITOPobs'].EtONO2/dfs['ITOPobs'].C2H6, dfs['ITOPobs'].sBuONO2/dfs['ITOPobs'].nC4H10, **ITOP_kwargs)
ax[0,0].scatter(dfs['TORCH1obs'].EtONO2/dfs['TORCH1obs'].C2H6, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[0,0].scatter(dfs['TORCH2obs'].EtONO2/dfs['TORCH2obs'].C2H6, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[0,0].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[0,0].set_ylabel('$EtONO_2$/$C_2H_6$')
ax[0,0].set_xlim(0,0.1)
# ax[0,0].set_xscale('log')
# ax[0,0].set_yscale('log')
# ax[0,0].legend(loc='upper right', frameon=True);

ax[0,1].scatter(dfs['NAMBLEXobs'].nPrONO2/dfs['NAMBLEXobs'].C3H8, dfs['NAMBLEXobs'].sBuONO2/dfs['NAMBLEXobs'].nC4H10, **NAMBLEX_kwargs)
ax[0,1].scatter(dfs['ITOPobs'].nPrONO2/dfs['ITOPobs'].C3H8, dfs['ITOPobs'].sBuONO2/dfs['ITOPobs'].nC4H10, **ITOP_kwargs)
ax[0,1].scatter(dfs['TORCH1obs'].nPrONO2/dfs['TORCH1obs'].C3H8, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[0,1].scatter(dfs['TORCH2obs'].nPrONO2/dfs['TORCH2obs'].C3H8, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[0,1].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[0,1].set_ylabel('$nPrONO_2$/$C_3H_8$')
ax[0,1].set_xlim(0,0.1)
# ax[0,1].set_xscale('log')
# ax[0,1].set_yscale('log')
# ax[0,1].legend(loc='upper right', frameon=True);

ax[0,2].scatter(dfs['NAMBLEXobs'].iPrONO2/dfs['NAMBLEXobs'].C3H8, dfs['NAMBLEXobs'].sBuONO2/dfs['NAMBLEXobs'].nC4H10, **NAMBLEX_kwargs)
ax[0,2].scatter(dfs['ITOPobs'].iPrONO2/dfs['ITOPobs'].C3H8, dfs['ITOPobs'].sBuONO2/dfs['ITOPobs'].nC4H10, **ITOP_kwargs)
ax[0,2].scatter(dfs['TORCH1obs'].iPrONO2/dfs['TORCH1obs'].C3H8, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[0,2].scatter(dfs['TORCH2obs'].iPrONO2/dfs['TORCH2obs'].C3H8, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[0,2].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[0,2].set_ylabel('$iPrONO_2$/$C_3H_8$')
ax[0,2].set_xlim(0,0.1)
# ax[0,2].set_xscale('log')
# ax[0,2].set_yscale('log')
ax[0,2].legend(loc='upper right', frameon=True);

ax[1,0].scatter(dfs['TORCH1obs'].nBuONO2/dfs['TORCH1obs'].nC4H10, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[1,0].scatter(dfs['TORCH2obs'].nBuONO2/dfs['TORCH2obs'].nC4H10, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[1,0].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[1,0].set_ylabel('$nBuONO_2$/$nC_4H_{10}$')
ax[1,0].set_xlim(0,0.1)
# ax[1,0].set_xscale('log')
# ax[1,0].set_yscale('log')
# ax[1,0].legend(loc='upper right', frameon=True);

ax[1,1].scatter(dfs['ITOPobs'].nPeBCONO2/dfs['ITOPobs'].nC5H12, dfs['ITOPobs'].sBuONO2/dfs['ITOPobs'].nC4H10, **ITOP_kwargs)
ax[1,1].scatter(dfs['TORCH1obs'].nPeBCONO2/dfs['TORCH1obs'].nC5H12, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[1,1].scatter(dfs['TORCH2obs'].nPeBCONO2/dfs['TORCH2obs'].nC5H12, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[1,1].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[1,1].set_ylabel('($nPeBONO_2+nPeCONO_2$)/$nC_5H_{12}$')
ax[1,1].set_xlim(0,3.5)
# ax[1,1].set_xscale('log')
# ax[1,1].set_yscale('log')
# ax[1,1].legend(loc='upper right', frameon=True);

ax[1,2].scatter(dfs['ITOPobs'].iPeBONO2/dfs['ITOPobs'].nC5H12, dfs['ITOPobs'].sBuONO2/dfs['ITOPobs'].nC4H10, **ITOP_kwargs)
ax[1,2].scatter(dfs['TORCH1obs'].iPeBONO2/dfs['TORCH1obs'].nC5H12, dfs['TORCH1obs'].sBuONO2/dfs['TORCH1obs'].nC4H10, **TORCH1_kwargs)
ax[1,2].scatter(dfs['TORCH2obs'].iPeBONO2/dfs['TORCH2obs'].nC5H12, dfs['TORCH2obs'].sBuONO2/dfs['TORCH2obs'].nC4H10, **TORCH2_kwargs)
ax[1,2].set_xlabel('$sBuONO_2$/$nC_4H_{10}$')
ax[1,2].set_ylabel('$iPeBONO_2$/$nC_5H_{12}$')
ax[1,2].set_xlim(0,3.5);
# ax[1,2].set_xscale('log')
# ax[1,2].set_yscale('log')
# ax[1,2].legend(loc='upper right', frameon=True);
```
