```python
import os
import netCDF4
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
%matplotlib inline
```

```python
path_oxbuds = '/local/mwe14avu/UEA/PhD/models/external_data/OXBUDS_emissions'
```

```python
_emis_nbut = Dataset(os.path.join(path_oxbuds,'n-butane_emissions.nc'))
_emis_ibut = Dataset(os.path.join(path_oxbuds,'i-butane_emissions.nc'))
_emis_npen = Dataset(os.path.join(path_oxbuds,'n-pentane_emissions.nc'))
_emis_ipen = Dataset(os.path.join(path_oxbuds,'i-pentane_emissions.nc'))
```

```python
lons = _emis_nbut.variables['longitude'][:]
lats = _emis_nbut.variables['latitude'][:]
month = _emis_nbut.variables['time'][:]
emis_nbut = _emis_nbut.variables['emiss_nbut'][:]
emis_ibut = _emis_ibut.variables['emiss_ibut'][:]
emis_npen = _emis_npen.variables['emiss_npen'][:]
emis_ipen = _emis_ipen.variables['emiss_ibut'][:]
_emis_nbut.close()
_emis_ibut.close()
_emis_npen.close()
_emis_ipen.close()
```

```python
# Year 2000 means
print('OXBUDS global mean emission [kg m-2 s-1] for year 2000',
      'n-butane: {_a:.2e}'.format(_a=emis_nbut[-13:-1, ...].mean()), 
      'i-butane: {_b:.2e}'.format(_b=emis_ibut[-13:-1, ...].mean()), 
      'n-pentane: {_c:.2e}'.format(_c=emis_npen[-13:-1, ...].mean()), 
      'i-pentane: {_d:.2e}'.format(_d=emis_ipen[-13:-1, ...].mean()), sep='\n')
```

```python
# Year 1990 means
print('OXBUDS global mean emission [kg m-2 s-1] for year 1990',
      'n-butane: {_a:.2e}'.format(_a=emis_nbut[-26:-14, ...].mean()), 
      'i-butane: {_b:.2e}'.format(_b=emis_ibut[-26:-14, ...].mean()), 
      'n-pentane: {_c:.2e}'.format(_c=emis_npen[-26:-14, ...].mean()), 
      'i-pentane: {_d:.2e}'.format(_d=emis_ipen[-26:-14, ...].mean()), sep='\n')
```

```python
# Plot
fig, ax = plt.subplots(subplot_kw=dict(projection=ccrs.PlateCarree()))
p = ax.contourf(lons, lats, emis_nbut[-1, ...], levels=np.logspace(-12, -10, 10), cmap='magma_r', transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_global()
fig.colorbar(p, orientation='horizontal')
plt.title('n-butane')
```
