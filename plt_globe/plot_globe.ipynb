```python
import matplotlib.pyplot as plt
import numpy as np
import cartopy
import cartopy.crs as ccrs
```

```python
%matplotlib inline
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection=ccrs.Orthographic(-10, 30))

ax.add_feature(cartopy.feature.OCEAN, zorder=0, facecolor='#669999')
ax.add_feature(cartopy.feature.LAND, zorder=0, facecolor='#2D882D', edgecolor='black')

ax.set_global()
ax.gridlines(xlocs=np.arange(-180, 181, 10),
             ylocs=np.arange(-90, 91, 10), color='k')

fig.savefig('globe.svg', transparent=True)
fig.savefig('globe.eps', transparent=True)

# Color combos
#669999
#2D882D
```
