```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/local/mwe14avu/UEA/PhD/models/external_data/alkane_ratios/Baker_et_al_2008.txt'
```

```python
df = pd.read_csv(path, skiprows=1)
```

```python
print(df.C2H6.mean(), df.C3H8.mean(), df.nC4H10.mean(), df.iC4H10.mean(), df.nC5H12.mean(), df.iC5H12.mean())
```

```python
# Calculate mean alkane ratios relative to ethane, i.e. ethane ratio == 1.0
coef_C3H8 = df.C3H8.mean()*1/df.C2H6.mean()
coef_nC4H10 = df.nC4H10.mean()*1/df.C2H6.mean()
coef_iC4H10 = df.iC4H10.mean()*1/df.C2H6.mean()
coef_nC5H12 = df.nC5H12.mean()*1/df.C2H6.mean()
coef_iC5H12 = df.iC5H12.mean()*1/df.C2H6.mean()
print(('{:3.2f} '*5).format(coef_C3H8, coef_nC4H10, coef_iC4H10, coef_nC5H12, coef_iC5H12))
```

```python
ax = df.boxplot(column=['C2H6', 'C3H8', 'nC4H10', 'iC4H10', 'nC5H12', 'iC5H12'])
ax.set_ylabel('ppt')
```

```python
C2H6_mean_m_std = df.C2H6 - df.C2H6std
C2H6_mean_p_std = df.C2H6 + df.C2H6std
C3H8_mean_m_std = df.C3H8 - df.C3H8std
C3H8_mean_p_std = df.C3H8 + df.C3H8std
nC4H10_mean_m_std = df.nC4H10 - df.nC4H10std
nC4H10_mean_p_std = df.nC4H10 + df.nC4H10std
iC4H10_mean_m_std = df.iC4H10 - df.iC4H10std
iC4H10_mean_p_std = df.iC4H10 + df.iC4H10std
nC5H12_mean_m_std = df.nC5H12 - df.nC5H12std
nC5H12_mean_p_std = df.nC5H12 + df.nC5H12std
iC5H12_mean_m_std = df.iC5H12 - df.iC5H12std
iC5H12_mean_p_std = df.iC5H12 + df.iC5H12std
```

```python
plt.plot(C2H6_mean_m_std, 'bo', linestyle='dotted')
plt.plot(df.C2H6, 'ko')
plt.plot(C2H6_mean_p_std, 'ro', linestyle='dotted')
```
